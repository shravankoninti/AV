```
import pandas as pd                                                  # to import csv and for data manipulation
import matplotlib.pyplot as plt                                      # to plot graph
import seaborn as sns                                                # for intractve graphs
import numpy as np                                                   # for linear algebra

import sweetviz
import warnings
warnings.filterwarnings('ignore')
```

```
!ls '/content/drive/My Drive/AV_data/MLP_Starter_Hack'
```

```
file = r'/content/drive/My Drive/AV_data/MLP_Starter_Hack/'
sub_df = pd.read_csv(file+'sample_submission.csv')
```

```
sub_df.head()
```

```
train_df = pd.read_csv(file+'train.csv')
train_df.head()
```

```
train_df.dtypes
```

```
test_df = pd.read_csv(file+'test.csv')
test_df.head()
```

```
train_df['trainee_engagement_rating'] = train_df['trainee_engagement_rating'].astype(str)
test_df['trainee_engagement_rating'] = test_df['trainee_engagement_rating'].astype(str)
```

### sweetviz

```
train_report = sweetviz.analyze([train_df, "Train"], target_feat = 'is_pass')
```

```
train_report.show_html('Train_Report.html')
```

```
print(train_df.shape, test_df.shape)
```

```
test_report = sweetviz.analyze([test_df, "Test"])
```

```
test_report.show_html('Test_Report.html')
```

### Compare two dataframes

```
train_test_report = sweetviz.compare([train_df, "Train"] , [test_df, "Test"], "is_pass")
```

```
train_test_report.show_html('Train_Test_Report.html')
```
